```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

matches=pd.read_csv('odi_Matches_Data.csv')
matches.dropna(subset=['Match Winner','Match Result Text'],inplace=True)

matches['results']=matches['Match Result Text'].str.strip().str.lower()

matches['chasing_win'] = matches['results'].str.contains('wickets')

chasing_matches=matches[matches['chasing_win']==True]
chasing_winners=chasing_matches['Match Winner']
chasing_counts=chasing_winners.value_counts()
chasing_df=chasing_counts.reset_index()
chasing_df.columns=['Team','Chasing Wins']

matches['defending_win'] = matches['results'].str.contains('runs')
defending_matches = matches[matches['defending_win']==True]
defending_winners = defending_matches['Match Winner']
defending_counts = defending_winners.value_counts()
defending_df = defending_counts.reset_index()
defending_df.columns = ['Team','Defending Wins']

final_df = pd.merge(chasing_df, defending_df,on='Team', how='outer')
final_df['Total Wins']=final_df['Chasing Wins']+final_df['Defending Wins']
final_df = final_df.dropna(subset=['Total Wins'])

final_df=final_df.sort_values(by='Total Wins',ascending=False)
final_df.reset_index(drop=True,inplace=True)

final_df



```

```python
final_df.reset_index(inplace=True)

plt.figure(figsize=(12, 6))
sns.barplot(data=final_df, x='Team', y='Total Wins', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Total Wins by Team')
plt.xlabel('Team')
plt.ylabel('Total Wins')
plt.tight_layout()
plt.show()
```
