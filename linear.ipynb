```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

matches=pd.read_csv('matches.csv')
matches['winner']=matches['winner'].replace({
    'Delhi Daredevils':'Delhi Capitals'
})
matches['winner']=matches['winner'].replace({
    'Rising Pune Supergiant':'Rising Pune Supergiants'
})
matches['winner']=matches['winner'].replace({
    'Gujarat Lions':'Gujarat Titans'
})
matches['winner']=matches['winner'].replace({
    'Kings XI Punjab':'Punjab Kings'
})

matches['winner']=matches['winner'].replace({
    'Royal Challengers Bengaluru':'Royal Challengers Bangalore'
})

matches.dropna(subset=['winner','result'],inplace=True)
matches['result']=matches['result'].str.strip().str.lower()

##chasing 

matches['chasing_win']=matches['result'].str.endswith('wickets')
chasing_matches=matches[matches['chasing_win']==True]
chasing_winners=chasing_matches['winner']
chasing_counts=chasing_winners.value_counts()
chasing_df=chasing_counts.reset_index()
chasing_df.columns=['Team','Chasing Wins']

## defending 

matches['defending_win'] = matches['result'].str.endswith('runs')
defending_matches = matches[matches['defending_win']==True]
defending_winners = defending_matches['winner']
defending_counts = defending_winners.value_counts()
defending_df = defending_counts.reset_index()
defending_df.columns = ['Team','Defending Wins']

##merge

final_df = pd.merge(chasing_df, defending_df,on='Team', how='outer')
final_df['Total Wins']=final_df['Chasing Wins']+final_df['Defending Wins']

active_teams=[
    'Chennai Super Kings',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Sunrisers Hyderabad',
    'Rajasthan Royals',
    'Punjab Kings',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
##active teams 

final_df.reset_index(inplace=True)
final_df['Team']=final_df['Team'].apply(lambda x: x + '*' if x in active_teams else x)

##sorting

final_df=final_df.sort_values(by='Total Wins',ascending=False)
final_df.set_index('Team', inplace=True)

```

```python
plt.figure(figsize=(10, 6))
plt.barh(final_df.index, final_df['Chasing Wins'], color='skyblue', label='Chasing Wins')
plt.barh(final_df.index, final_df['Defending Wins'],left=final_df['Chasing Wins'], color='orange', label='Defending Wins')
plt.title('IPL Teams: Chasing vs Defending Wins',fontsize=16)
plt.xlabel('Number of Wins',fontsize=12)
plt.ylabel('Teams(* Denotes Active Teams)',fontsize=12)
plt.legend()
```
